### Description of a Crop Recommendation System with Soil Data
A crop recommendation system helps farmers select the best crops to grow based on the specific properties of their soil. This system uses soil characteristics and environmental factors to determine the crops that are most likely to thrive. Recommendations are provided to improve crop yield, optimize resource use, and ensure sustainable farming practices.


### Core Components for Recommendations
The system should consider the following soil parameters and external factors to make accurate recommendations:

1. Soil Nutrients:

    - Nitrogen (N): Promotes leafy growth; ideal for crops like spinach, lettuce, and wheat.
    - Phosphorus (P): Essential for root development; crucial for legumes, peas, and root vegetables like carrots.
    - Potassium (K): Enhances disease resistance and fruit quality; important for fruiting plants like tomatoes, bananas, and potatoes.

2. Soil pH:

    - Indicates soil acidity or alkalinity.
    - Neutral pH (6.5-7.5) supports most crops like rice, wheat, and maize.
    - Acidic soil (<6.5) favors crops like tea and coffee.
    - Alkaline soil (>7.5) supports crops like barley and asparagus.

3. Organic Matter:

    - High organic content improves water retention and nutrient availability.
    - Crops like vegetables and fruits benefit from rich organic matter.

4. Moisture Level:

    - Determines irrigation needs and crop suitability.
    - High moisture crops: Paddy, sugarcane.
    - Low moisture crops: Millet, sunflower.

5. Temperature:

    - Warm crops: Maize, rice, and cotton.
    - Cool crops: Wheat, barley, and cabbage.

6. Rainfall:

    - Rain-fed crops (e.g., rice) thrive in high rainfall areas.
    - Drought-resistant crops (e.g., millets) perform well in low-rainfall zones.

7. Geographical Factors:

    - Altitude, latitude, and local climate conditions.
    - Example: Coffee grows well in high altitudes, while coconut thrives in coastal regions.]

**Complete description about columns in dataset:**
1. Temperature
 - Label: Temperature
 - Description: Represents the range of temperatures (in °C) suitable for crop growth, as it impacts germination and plant development.
 - Example: 26.00 - 28.00°C indicates moderate warmth, ideal for crops like rice or maize.
2. Humidity
 - Label: Humidity
 - Description: Represents the percentage of atmospheric moisture, influencing plant water needs and drought resistance.
 - Example: 56.01 - 60.01% is suitable for most crops like wheat, rice, and vegetables.
3. Moisture
 - Label: Moisture
 - Description: Indicates the percentage of water content in the soil, essential for crop hydration and root absorption.
 - Example: 35.00 - 40.00% is suitable for water-demanding crops like sugarcane or paddy.
4. Soil Type
 - Label: Soil Type
 - Description: Categorizes soil based on texture and composition (e.g., clayey, black, or other), which affects water retention and nutrient availability.
 - Example: Black soil is ideal for crops like cotton and sugarcane.
5. Crop Type
 - Label: Crop Type
 - Description: Specifies the type of crop being recommended or grown, categorized by the plant's purpose or growth pattern.
 - Example: Sugarcane, maize, or other crop types.
6. Nitrogen
 - Label: Nitrogen
 - Description: Represents the nitrogen content in the soil (in mg/kg or percentage), critical for leafy growth and overall plant health.
 - Example: 13.80 - 18.40 mg/kg is moderate and suitable for leafy vegetables like spinach.
7. Potassium
 - Label: Potassium
 - Description: Represents the potassium content in the soil (in mg/kg or percentage), essential for plant immunity and fruit quality.
 - Example: 2.30 - 4.60 mg/kg is suitable for fruiting crops like tomatoes and bananas.
8. Phosphorus
 - Label: Phosphorus
 - Description: Represents the phosphorus content in the soil (in mg/kg or percentage), necessary for root development and flowering.
 - Example: 18.40 - 23.00 mg/kg is suitable for legumes like peas and beans.
9. Fertilizer Name
 - Label: Fertilizer Name
 - Description: Specifies the recommended type of fertilizer to optimize crop growth based on soil deficiencies.
 - Example: Urea for nitrogen-deficient soil or DAP (Diammonium Phosphate) for phosphorus deficiency.

First I need to convert categorical variables into numerical values.
There are different ways to convert.

1. **One Hot Encoding** - create binary column for each category (for handling categorical variables with multiple levels, eg: red, green, blue)
2. **Label Encoding** - assign a unique integer to each category (for handling values with *order*, eg: small,  medium, large)
3. **Dummy Encoding/ Indicator encoding** - similar to one hot, but doesn't remove any columns (for working with categorical variables with many levels, eg: 10 different colors)
4. **Binning** - group numerical values into bin/ ranges (for handling large number of unique values, eg: age groups like young, middle-aged, old)
5. **Count Encoding** - counting number of occurences in each category (for handling categorical variables with multiple levels, eg: 10 rows of apples as 'apple | 10' in each apple row)
6. **Frequency Encoding** - representing each category as a proportion of occurances of that category in dataset (for categorical variables with many levels)
7. **Target Encoding** - representing each category as the mean of the target variable for that category (useful where data is highly imbalanced)

Since there are only 5 soil types and they doesn't have any order, I choose **One Hot Encoding** here. 

```python
df = pd.get_dummies(df, columns=['Soil Type'], drop_first=False)
```

Now we have to convert Crop type column which is a categorical column to numerical column. 

The technique here is use is **Label Encoding**. 

*Why is Label Encoding Okay for Target (Y)?*
 - Label Encoding is problematic when used as an input (X), because ML models might assume 0 < 1 < 2 ... < 10, even if there's no real ranking.
 - But when Label Encoding is used for the target variable (Y), it’s just assigning unique class labels, not ranking them.

    Example:
    If we encode crops like this:
    - Wheat → 0
    - Rice → 1
    - Corn → 2

A classification model (like Decision Tree, Random Forest, or Neural Network) will treat them as separate categories, not ordered numbers.

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['Crop Type'] = label_encoder.fit_transform(df['Crop Type'])

# Check encoding mapping
print(dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))))
```

What is *Standardization*?

Standardization is a technique used to scale numerical features so they have zero mean (𝜇 = 0) and unit variance (𝜎² = 1).

This transformation helps machine learning models work better when features have very different ranges.

When to Standardize?

 - ✅ If using distance-based models (SVM, KNN, Logistic Regression, Neural Networks) → Yes, standardize!
 - ❌ If using tree-based models (Decision Trees, Random Forest, XGBoost) → No need to standardize!

Let's split the data set for training. Here we need to divide data set  into features (X) and target (Y).

```python
from sklearn.model_selection import train_test_split

# Define x (features) and Y (target)
x = df.drop(columns=['Crop Type'])
y = df['Crop Type']

# Split data into 80% training, 20% tesing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

Why use `random_state`?

 - If not set, `train_test_split` will shuffle randomly every time, giving different results each time you run it.
 - Setting `random_state=42` (42 is random number) ensures the same split every time, making your experiments reproducible. (ensures consistent results every time we run the code)

#### Additional: 
If we have class imbalances, for example in 1000 sample of crops 700 are wheat, 200 are rice and 100 are corn, we can use the argument `stratify=y`.

Why use `stratify=y`?
 - Prevents class imbalance issues
 - Ensures fair distribution of all crop types
 - Helps models generalize better


## Common Classification Models

Main Types:

**1. Eager Learners**

First build a model from the training dataset before making any prediction on future dataset. They spend more time on training the model to have a better generalization, but they take less time to make predictions.

eg: Logistic Regresstion, Support Vector Machines, Decision Trees, Artificial Neural Network (ANN)

**2. Lazy Learners/ Instance based Leaners**

Don't create any model from training data, they just memorize training data and each time there is a need to make a prediction, they search for the nearest neighbour from whole training data. Therefore they are vey slow on making predictions. 

eg: K-Nearest Neighbour(KNN), Case based reasoning

 Okay, now let's train the model.
 I am going to use **Decision Tree** for this, because it's good for categorical prediction. 

```python
from sklearn.tree import DecisionTreeClassifier

# train model
model = DecisionTreeClassifier(class_weight="balanced", max_depth=10, min_samples_split=5, min_samples_leaf=2, random_state=42)
model.fit(x_train, y_train)
```

```python
# Predict on test data
y_predict = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_predict)
print('Model Accuracy: ', accuracy)
```

Initially I got 0.0875, then I use class balancing in Decision Tree (`class_weight="balanced", max_depth=10, min_samples_split=5, min_samples_leaf=2, random_state=42`)

 - max_depth=10 → Prevents overfitting
 - min_samples_split=5 → Node must have at least 5 samples to split
 - min_samples_leaf=2 → Minimum samples per leaf node

 After these methods, now accuracy is 0.103125 (10%) which is still bad.

```python
from sklearn.ensemble import RandomForestClassifier

model1 = RandomForestClassifier(class_weight="balanced", n_estimators=100, max_depth=10, random_state=42)
model1.fit(x_train, y_train)

y_predict1 = model1.predict(x_test)
accuracy1 = accuracy_score(y_test, y_predict1)
print('Random Forest Model Accuracy: ', accuracy1)
```

```python
# Hyperparameter tuning for Decision Tree
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

param_grid_dt = {
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}

dt = DecisionTreeClassifier(class_weight='balanced', random_state=42)
grid_search_dt = GridSearchCV(dt, param_grid_dt, cv=5, scoring='accuracy')
grid_search_dt.fit(x_train, y_train)

print("Best Decision Tree parameters:", grid_search_dt.best_params_)
best_dt = grid_search_dt.best_estimator_
dt_accuracy = accuracy_score(y_test, best_dt.predict(x_test))
print('Improved Decision Tree Accuracy:', dt_accuracy)

# Hyperparameter tuning for Random Forest
from sklearn.ensemble import RandomForestClassifier

param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'class_weight': ['balanced', 'balanced_subsample']
}

rf = RandomForestClassifier(random_state=42)
grid_search_rf = GridSearchCV(rf, param_grid_rf, cv=5, scoring='accuracy')
grid_search_rf.fit(x_train, y_train)

print("Best Random Forest parameters:", grid_search_rf.best_params_)
best_rf = grid_search_rf.best_estimator_
rf_accuracy = accuracy_score(y_test, best_rf.predict(x_test))
print('Improved Random Forest Accuracy:', rf_accuracy)
```

Analyze Feature Importance

```python
import matplotlib.pyplot as plt
import numpy as np

# For Random Forest
best_rf = grid_search_rf.best_estimator_
feature_importances = best_rf.feature_importances_
feature_names = x.columns
indices = np.argsort(feature_importances)[::-1]

plt.figure(figsize=(12, 8))
plt.title('Feature Importance')
plt.bar(range(x.shape[1]), feature_importances[indices], align='center')
plt.xticks(range(x.shape[1]), [feature_names[i] for i in indices], rotation=90)
plt.tight_layout()
plt.show()

# Keep only important features (those with importance > threshold)
threshold = 0.01  # Adjust this threshold based on your analysis
important_indices = [i for i, imp in enumerate(feature_importances) if imp > threshold]
important_features = x.columns[important_indices]

print("Important features:", important_features)

# Retrain with important features only
x_important = x[important_features]
x_train_imp, x_test_imp, y_train, y_test = train_test_split(x_important, y, test_size=0.2, random_state=42)

best_rf.fit(x_train_imp, y_train)
rf_pruned_accuracy = accuracy_score(y_test, best_rf.predict(x_test_imp))
print('Random Forest Accuracy with feature selection:', rf_pruned_accuracy)
```
